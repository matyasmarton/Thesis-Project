```python
# import the dependencies
from __future__ import print_function
import cv2
import face_recognition
import numpy as np
# this string will be used at the export to write to the FaceDetected folder
path = './FaceDetected/'

for file in files:
#read the files

  img =  cv2.imread(file,1)
#convert files to numpy array

  image = np.array(img)
#detect faces, second argument is for deteailed search, higher the number, the better it finds smaller faces

  face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=2, model="cnn")
 #if there is one or more faces detected write file to selected path 

  if len(face_locations) > 0:
    cv2.imwrite(os.path.join(path, f'{file}'), img)
    print(path+file)

# if no faces were detected print the below image
  else:
    print('not saved'+file)
```
